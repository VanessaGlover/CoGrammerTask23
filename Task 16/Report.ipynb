```python
# Create a DataFrame with balance.txt

import pandas as pd
df = pd.read_csv('balance.txt', delim_whitespace=True)

balance_data = { 
    'Ethnicity': ['Asian', 'White', 'Black', 'Hispanic', 'White', 'Black'],
    'Marital_Status': ['Married', 'Single', 'Single', 'Married', 'Married', 'Single'],
    'Income': [520000, 120000, 55000, 90000, 5000, 28000],
    'Balance': [2000, 1000, 2800, 12000, 33500, 2200],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Female', 'Male'],
    'Number_of_Cards': [2, 3, 1, 2, 2, 1]
    }

df = pd.DataFrame(balance_data)

# Comparing the average income based on ethnicity
average_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print("Average Income Based on Ethnicity:")
for ethnicity, income in average_income_by_ethnicity.items():
    print(f"Ethnicity: {ethnicity} | Income: {income:.2f}")
print()

average_balance_by_marital_status = df.groupby('Marital_Status')['Balance'].mean()
print("Average Balance Based on Marital Status:")
for status, balance in average_balance_by_marital_status.items():
    print(f"{status}: {balance:.2f}")
print()

# What is the highest income in our dataset?
highest_income = df['Income'].max()
print("Highest income in the Dataset is ", highest_income)
print()

# What is the lowest income in our dataset?
lowest_income = df['Income'].min()
print("Lowest income in the Dataset is ", lowest_income)
print()

# How many cards do we have recorded in our dataset?
total_cards_recorded = df['Number_of_Cards'].sum()
print("Total Number of Cards Recorded in the Dataset is ", total_cards_recorded)
print()

# How many females do we have information for vs how many males?
gender_counts = df['Gender'].value_counts()
for gender, count in gender_counts.items():
    print(f"{gender}: {count}")
```
