```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Read the CSV file into a DataFrame
df = pd.read_csv('insurance.csv')
df.head()

# Create a scatter plot of Age vs Charges
df.plot.scatter( x= 'age', y= 'charges')
plt.title('Scatter Plot of Age vs Charges')
plt.show

# Extract the 'age' and 'charges' columns from the DataFrame
ages = df[['age']]  
charges = df['charges']  

model = LinearRegression()
model.fit(ages, charges)

# Make predictions on the data
predicted_charges = model.predict(ages)

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])

print("Predicted charges:", predicted_charges)

plt.figure(figsize=(10, 6))
plt.scatter(ages, charges, color='blue', label='Actual data')
plt.plot(ages, predicted_charges, color='red', label='Best fit line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Age vs Charges with Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```
